# Encoding 

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
X_encoded = X.copy()
```

```python
X_encoded[X_encoded.columns].nunique()
```

```python
X_encoded
```

### Applying One-Hot Encoding on Features with â‰¤5 unique values
### Applying Label Encoding on Features with >5 unique values

```python
for col in categorical_cols:
    unique_vals = X_encoded[col].nunique()
    
    if unique_vals <= 5:
        # One-Hot Encoding
        one_hot = pd.get_dummies(X_encoded[col], prefix=col)
        X_encoded = X_encoded.drop(columns=[col])      #drops the original categorical column from the dataset
        X_encoded = pd.concat([X_encoded, one_hot], axis=1)   #Combines the original DataFrame with the new one-hot encoded columns 
    else:
        # Label Encoding
        X_encoded[col] = label_encoder.fit_transform(X_encoded[col])
```

# Scaling 

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Initialize the scaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)
```

```python
# convert back to DataFrame for readability
X_scaled = pd.DataFrame(X_scaled, columns=X_encoded.columns)
```

```python
X_encoded.head()
```

```python
X_scaled.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
#Initialize and train SVM model ()
svm_model = SVC()
svm_model.fit(X_train, y_train)
```

```python
# Predictions
y_pred = svm_model.predict(X_test)
```

```python
# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
print(svm_model.score(X_train,y_train))
print(svm_model.score(X_test,y_test))
```

# Hyperparameter Tuning 

```python
import warnings
warnings.filterwarnings('ignore')
kernels = ['linear', 'rbf', 'poly', 'sigmoid']
results = {}

for kernel in kernels:
    print(f"\n Kernel: {kernel}")
    model = SVC(kernel=kernel)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {acc:.4f}")
    print("Classification Report:\n", classification_report(y_test, y_pred))
    print("Training accuracy", model.score(X_train,y_train))
    print("Testing accuracy", model.score(X_test,y_test))
    
    # Store results
    results[kernel] = acc
```

# Hyperparameter Tuning using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto'],
    'degree': [2, 3, 4]  # Only relevant for 'poly' kernel
}
```

```python
grid = GridSearchCV(
    estimator=SVC(),
    param_grid=param_grid,
    scoring='accuracy',
    cv=5,                # 5-fold cross-validation
    verbose=2,
    n_jobs=-1            # Use all available cores
)
```

```python
# Fit on training data
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
# Best parameters and score
print("Best Parameters Found:", grid.best_params_)
print(f"Best Accuracy from Grid Search: {grid.best_score_:.4f}")
```

```python
# Evaluate on test set
best_model = grid.best_estimator_
y_pred = best_model.predict(X_test)
```

```python
print("\nTest Set Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
print("Training accuracy", best_model.score(X_train,y_train))
print("Testing accuracy", best_model.score(X_test,y_test))    
```

# Strength and weakness for Mushroom Dataset
### ðŸŸ¢ Strengths
1. Handles Non-linear Decision Boundaries Well
2. Effective with High-dimensional Feature Space
3. Robust to Imbalanced Feature Scales (post-scaling)
4. Good Performance with Limited Data (~2000 samples)
5. Clear Visual Separation in PCA Plot
### ðŸ”´ Weaknesses
1. Overfitting Risk (High Train, Lower Test Accuracy)
2. Sensitivity to Parameter Tuning
3. Scaling Is Mandatory
4. Lack of Interpretability
5. Computationally Intensive on Larger Datasets
### Conclusion
Support Vector Machine (SVM), particularly with the polynomial kernel, proved to be a strong performer for the mushroom dataset, effectively handling its high-dimensional, categorical nature. However, careful tuning and scaling were essential to avoid overfitting and ensure generalization.
While SVM lacks interpretability and may not scale well to much larger datasets, it is a powerful choice for this moderately sized, structured classification task.

## Practical Implications of SVM in Real-World Classification Tasks
1. Effective in High-Dimensional Spaces
2. Useful When Classes Are Well-Separated
3. Robust Generalization with Proper Tuning
4. SVM Is Kernel-Driven â€” Adaptable to Various Data Structures
5. Good for Sensitive or Regulated Domains
### Limitations with Real-World Implications
1. Not Ideal for Very Large Datasets
2. Poor Interpretability
3. Sensitive to Feature Scaling and Noisy Data
