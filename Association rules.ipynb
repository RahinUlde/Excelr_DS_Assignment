```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("Online retail.xlsx")
df
```

```python
#List comprehension
data = [i.split(',') for i in df['Column']]
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
#TransactionEncoder convertsitem lists into transaction data for frequent itemset
te = TransactionEncoder()
encoded_df = te.fit_transform(data)
```

```python
encoded_df
```

```python
te.columns_
```

```python
data = pd.DataFrame(encoded_df,columns=te.columns_)
data
```

```python
data.replace(True,1,inplace=True)
data.replace(False,0,inplace=True)
data
```

```python
df = data.drop('asparagus', axis=1)
```

```python
from mlxtend.frequent_patterns import association_rules,apriori
```

```python
scores = apriori(df, min_support = 0.5, use_colnames = True)
scores
```

```python
scores1 = apriori(df, min_support = 0.05, use_colnames = True)
scores1
```

```python
rules1 = association_rules(scores1, metric = "lift", min_threshold = 1.0)
rules1
```

```python
plt.scatter(rules1['support'],rules1['lift'],alpha=0.5)
plt.title('Support v/s Lift')
plt.xlabel('Support')
plt.ylabel('Lift')
plt.show()
```

```python
association_rules(scores1, metric = "lift", min_threshold = 0.2)
```

#### After building Apriori model on given data y setting min_support value as 0.05,0.04 or 0.03 we feel that the result are too few compared to the size of dataset.

```python
scores4 = apriori(df, min_support = 0.02, use_colnames = True)
scores4
```

#### Setting min_support=0.02 means that frequent itemsets appearing in at least 2% of the transactions are being considered.

```python
rules4 = association_rules(scores4, metric = "lift", min_threshold = 1.0)
rules4
```

```python
#to ignore repeating transaction (duplicates transaction)
rules4 = rules4.iloc[::2]
```

## Inferences from visualizations
#### The results are analyzed using scatter plots of Support vs Lift, leading to the following insights:

#### Higher min_support values (0.05, 0.04 and 0.03):
- Generated fewer association rules (very few points on the scatter plot).
- Limited the discovery of strong item associations.
- The lift values were mostly between 1.2 and 1.7, indicating only mild correlations.
- Many potentially useful rules were filtered out due to the high support threshold.

#### Lower min_support (0.02) - Best Performing Model:
- More association rules were generated, as seen in the denser scatter plot.
- A wider range of lift values (up to 2.2) was observed, indicating stronger associations between products.
- More frequent itemsets were identified while still maintaining meaningful support.
- This setting strikes a balance between support and confidence, making it the most effective for discovering insightful associations.

## Final Conclusion
From this analysis, min_support = 0.02 is the optimal choice for this dataset as it captures more meaningful associations without being too restrictive. Higher thresholds (0.05 and 0.04) eliminate many potentially valuable insights, making them less suitable.

# Interview Questions

#### 1.What is lift and why is it important in Association rules?
Answer:
Lift is a metric used in association rule to measure how much more likely two items are to be bought together compared to if they were bought independently.
Importance of lift in Association rule:
It is used to understand strength of relationship between items in dataset.
Lift is important because it helps businesses identify strong associations between products.
For example, if "spaghetti â†’ ground beef" has a high lift value, it means people who buy spaghetti are highly likely to buy ground beef too. This insight can be used for product bundling, store layout optimization, and targeted marketing strategies.

#### 2.What is support and Confidence. How do you calculate them?
Answer:
Support and Confidence are two key metrics in Association Rule Mining that help evaluate the strength of relationships between items in a dataset.
Support: It tells us how frequently an item or itemset appears in the dataset.
Formula for Support: Support(A â†’ B) = (Transactions containing A and B) / (Total No. transactions)
Confidence: It measures how likely item B is to be bought when item A is bought.
Formula for Support: Confidence(A â†’ B) = (Transactions containing A and B) / (Transactions containing A)


#### 3. What are some limitations or challenges of Association rules mining?
Answer:
Few of challenges and limitation in association rules mining are:
###### Choosing the Right Thresholds:
Selecting appropriate minimum support and confidence values is challenging because setting them too high may eliminate useful patterns, while setting them too low may generate too many insignificant rules.

###### Large Dataset Complexity:
Generating frequent itemsets requires scanning a large dataset multiple times, which can be computationally expensive and time-consuming.

###### High Number of Rules:
The algorithm can generate a massive number of association rules, making it difficult to identify the most useful ones.

Despite these challenges, Association Rule Mining remains a powerful technique for uncovering patterns in large datasets. To overcome its limitations, businesses use optimization techniques, better threshold selection, and domain knowledge to extract the most valuable insights.
