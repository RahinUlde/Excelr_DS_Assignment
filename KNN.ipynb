# Droping unwanted column

```python
df.drop(columns = 'animal name', inplace = True, axis=1)
```

## Dividing data into dependent and independent 

```python
X = df.drop(columns=["type"])  
y = df["type"] 
```

```python
X
```

# Splitting Data into training and testing sets

```python
from sklearn.model_selection import train_test_split
```

```python
#stratify keep class proporion (handle class imbalance)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

## Scaling because distance calculation is present in KNN

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
#the scaler learns how to scale, based only on the training data.
X_train_scaled = scaler.fit_transform(X_train) # fit ONLY on training data
#applies the same transformation to test data, using only the training-set parameters
X_test_scaled  = scaler.transform(X_test) # fit ONLY on testing data
```

```python
#print(scaler.fit_transform(X_train)) # fit ONLY on training data
#print(scaler.fit_transform(X_test)) # fit ONLY on testing data
```

```python
X_train_scaled
```

```python
X_test_scaled
```

# Model Building

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#weights="distance" helps with class imbalance
knn = KNeighborsClassifier(n_neighbors=5, weights="distance")
knn.fit(X_train_scaled, y_train)
```

```python
y_pred = knn.predict(X_test_scaled)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification report:\n", classification_report(y_test, y_pred))
print("\nConfusion matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
knn.score(X_train_scaled,y_train), knn.score(X_test_scaled,y_test)
```

## Hyperparameter Tuning

```python
train_acc = []
test_acc = []

for i in range(1,20):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_scaled,y_train)
    train_acc.append(knn.score(X_train_scaled,y_train))
    test_acc.append(knn.score(X_test_scaled,y_test))
```

```python
plt.plot(range(1,20), train_acc, color = 'blue', label = 'Training Accuracy')
plt.plot(range(1,20), test_acc, color = 'red', label = 'Testing Accuracy')
plt.xlabel('K-values')
plt.ylabel('Accuracy')
plt.title('K-value Selection',size = 30)
plt.legend()
```

After building the K-Nearest Neighbors (KNN) model with k = 5, the model achieved 100% accuracy on both the training and testing sets, indicating perfect classification on the given dataset.
To validate this choice, I performed hyperparameter tuning and found that k = 5 yielded the best results among the tested values, confirming it as the optimal value for k in this case.
This suggests that the selected features and preprocessing steps (including scaling and stratified splitting) effectively preserved class separability for the KNN algorithm

```python
import numpy as np
from sklearn.decomposition import PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)
# â”€â”€ 4. TRAIN KNN (kâ€¯=â€¯5, distanceâ€‘weighted) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
knn = KNeighborsClassifier(n_neighbors=5, weights="distance")
knn.fit(X_pca, y)
# â”€â”€ 5. CREATE A MESHGRID & PREDICT ON IT â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
x_min, x_max = X_pca[:, 0].min() - 1, X_pca[:, 0].max() + 1
y_min, y_max = X_pca[:, 1].min() - 1, X_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.05), np.arange(y_min, y_max, 0.05))
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
# â”€â”€ 6. PLOT DECISION BOUNDARIES â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.30)     # regions
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, edgecolor="k")  # data points
plt.title("KNN (kâ€¯=â€¯5) Decision Boundaries")
plt.xlabel("Principal ComponentÂ 1")
plt.ylabel("Principal ComponentÂ 2")
plt.tight_layout()
plt.show()
```

# Interview Question
#### 1. What are the key hyperparameters in KNN?
##### Ans. Key hyperparameters in KNN are 'n_neighbors' (number of neighbors), 'weights' (uniform or distance-based), and 'metric' (distance function like Euclidean or Manhattan).
#### 2. What distance metrics can be used in KNN?
##### Ans. Common distance metrics in KNN include 'euclidean', 'manhattan', 'minkowski' (default), and 'chebyshev', chosen using the 'metric' parameter.
