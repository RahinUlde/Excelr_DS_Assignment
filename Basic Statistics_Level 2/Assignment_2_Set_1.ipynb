```python
sns.boxplot(x)
plt.show()
```

```python
x=pd.Series(x)
Q1 = np.quantile(x,0.25)
Q3 = np.quantile(x,0.75)
med = np.median(x)
IQR = Q3 - Q1
upper_limit = Q3+(1.5*IQR)
lower_limit = Q1-(1.5*IQR)
```

```python
print('First Quantile=', Q1, 'Second Quantile=', med, 'Third Quantile=', Q3,
      'Inter-Quartile Range=', IQR, 'Upper Whisker=', upper_limit, 'Lower Whisker=', lower_limit)
Outliers = x[(x <= lower_limit) | (x >= upper_limit)]
print('The outlier in the boxplot:',Outliers)
```
