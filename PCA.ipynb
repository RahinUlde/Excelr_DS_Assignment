### PCA 

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pcs = pca.fit_transform(x)
```

```python
pcs
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)*100
```

```python
plt.figure(figsize=(5,3),dpi=200)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.axvline(x = 8, linestyle = '--', color = 'red')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance')
plt.title('Scree Plot', size = 20)
```

```python
pca.components_
```

```python
pca = PCA(n_components = 8)
pcs = pca.fit_transform(x)
pcs
```

### Clustering  

```python
from sklearn.cluster import KMeans
```

### Clustering on original data

```python
sns.scatterplot(x = df['Alcohol'], y = df['Ash'])
```

```python
km = KMeans(n_clusters = 6)
ypred = km.fit_predict(x)
ypred
```

```python
km.cluster_centers_
```

```python
plt.scatter(df['Alcohol'],df['Ash'], c = ypred, cmap='rainbow')
```

```python
#inertia is wcss value 
km.inertia_
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
wcss = []

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(x)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.axvline(x = 4, linestyle = '--', color = 'red')
plt.xlabel('K-Values')
plt.ylabel('WCSS')
plt.title('Elbow Graph')
```

```python
km2 = KMeans(n_clusters = 3)
ypred2 = km2.fit_predict(x)
ypred2
```

```python
#cluster centre when no. of cluster is 6
km2.cluster_centers_
```

```python
plt.scatter(df['Alcohol'],df['Ash'], c = ypred2, cmap='rainbow')
```

### Clustering on PCA data

```python
x2 = data.iloc[:,:-1]
```

```python
x2
```

```python
plt.scatter(data['PC1'],data['PC6'])
```

```python
km3 = KMeans(n_clusters = 4)
ypred3 = km3.fit_predict(x2)
ypred3
```

```python
km3.cluster_centers_
```

```python
plt.scatter(data['PC1'],data['PC6'], c = ypred3)
```

```python
km4 = KMeans(n_clusters = 3)
ypred4 = km4.fit_predict(x2)
ypred4
```

```python
plt.scatter(data['PC1'],data['PC6'], c = ypred4)
```

As we can see overlaping and not so good clustering when clustering is done by using original data. Applying Principal Component Analysis (PCA) before K-Means clustering improved the clustering performance compared to using the original data. By reducing dimensionality and eliminating noise, PCA helped in achieving better cluster separation and more meaningful groupings. This demonstrates that PCA can enhance clustering effectiveness by capturing the most significant variance in the data while reducing computational complexity.
